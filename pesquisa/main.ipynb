```python
import pandas as pd
import matplotlib.pyplot as pyplot
```

```python
#TABELA:DADOS DO CSV FILTRADO POR HORARIO
dataframe = pd.read_csv("Dados_PRF_2022.csv", sep = ';', encoding = 'iso-8859-1')
dataframe.sort_values(by = ['horario'], ascending = False)
```

```python
#CONSOLE:QUANTIDADE DE ACIDENTES DE CADA TIPO
contagem_causas  = dataframe['causa_acidente'].value_counts()
print(contagem_causas)
nomeCausa = contagem_causas.idxmax()
qtdCausa = contagem_causas.max()
print(f'Nome da causa: {nomeCausa} - Quantidade de acidentes: {qtdCausa}')
print(dataframe.columns)
#site para o relatorio:https://agenciabrasil.ebc.com.br/geral/noticia/2023-09/acidentes-por-reacao-tardia-do-condutor-crescem-28-em-2022#:~:text=%E2%80%9CA%20rea%C3%A7%C3%A3o%20tardia%2C%20ineficiente%20ou,viagem%E2%80%9D%2C%20explicou%20em%20comunicado.
#############
a  = dataframe['condicao_metereologica'].value_counts()
print(a)
b = a.idxmax()
c = a.max()
print(f'Nome da causa: {b} - Quantidade de acidentes: {c}')
```

```python
#CONSOLE:QUANTIDADE DE ACIDENTES POR DIAS DA SEMANA
contagem_dia_semana = dataframe['dia_semana'].value_counts().sort_index()
print(contagem_dia_semana)
qualDia = contagem_dia_semana.idxmax()
quantos = contagem_dia_semana.max()
print(f'Dia da semana: {qualDia} - Quantidade de acidentes: {quantos}')
```

```python
#GRÁFICO:QUANTIDADE DE ACIDENTES POR DIAS DA SEMANA
contagem_dia_semana = dataframe['dia_semana'].value_counts().sort_index()
pyplot.figure(figsize=(10, 6))
contagem_dia_semana.plot(kind='bar', color='skyblue')
pyplot.xlabel('Dia da Semana')
pyplot.ylabel('Quantidade de Acidentes')
pyplot.title('Quantidade de Acidentes por Dia da Semana')
pyplot.xticks(rotation=45)
pyplot.show()
#site para o relatorio:https://www.correiobraziliense.com.br/app/noticia/cidades/2011/12/20/interna_cidadesdf,283492/maior-parte-dos-acidentes-ocorre-entre-o-fim-da-tarde-e-meia-noite.shtml
#finais de semana são os dias que tem mais casos, provavelmente por conta que as pessoas não estão trabalhando
```

```python
#CONSOLE:QUANTIDADE DE ACIDENTES POR INTERVALO DE HORÁRIO
intervalos = ['00:00 - 03:00', '03:00 - 06:00', '06:00 - 09:00', '09:00 - 12:00',
              '12:00 - 15:00', '15:00 - 18:00', '18:00 - 21:00', '21:00 - 00:00']

# Criar uma função para atribuir cada horário a um intervalo
def atribuir_intervalo(horario):
    if '00:00' <= horario < '03:00':
        return '00:00 - 03:00'
    elif '03:00' <= horario < '06:00':
        return '03:00 - 06:00'
    elif '06:00' <= horario < '09:00':
        return '06:00 - 09:00'
    elif '09:00' <= horario < '12:00':
        return '09:00 - 12:00'
    elif '12:00' <= horario < '15:00':
        return '12:00 - 15:00'
    elif '15:00' <= horario < '18:00':
        return '15:00 - 18:00'
    elif '18:00' <= horario < '21:00':
        return '18:00 - 21:00'
    else:
        return '21:00 - 00:00'

# Aplicar a função de atribuição de intervalo à coluna 'horario'
dataframe['intervalo'] = dataframe['horario'].apply(atribuir_intervalo)

# Agrupar por intervalo e contar a quantidade de acidentes em cada intervalo
contagem_por_intervalo = dataframe['intervalo'].value_counts().reindex(intervalos, fill_value=0)

# Exibir o resultado
print(contagem_por_intervalo)
#site para o relatorio:https://www.correiobraziliense.com.br/app/noticia/cidades/2011/12/20/interna_cidadesdf,283492/maior-parte-dos-acidentes-ocorre-entre-o-fim-da-tarde-e-meia-noite.shtml
#devido ao horario de pico, nesses intervalos normalmente as pessoas vão trabalhar
```

```python
#GRÁFICO:QUANTIDADE DE ACIDENTES POR INTERVALO DE TEMPO

pyplot.figure(figsize=(10, 6))
pyplot.plot(intervalos, contagem_por_intervalo, marker='o', linestyle='-', color='b')
pyplot.xlabel('Intervalo de Tempo')
pyplot.ylabel('Quantidade de Acidentes')
pyplot.title('Quantidade de Acidentes por Intervalo de Tempo')
pyplot.xticks(rotation=45)
pyplot.grid(True)
pyplot.show()
#site para o relatorio:https://www.correiobraziliense.com.br/app/noticia/cidades/2011/12/20/interna_cidadesdf,283492/maior-parte-dos-acidentes-ocorre-entre-o-fim-da-tarde-e-meia-noite.shtml
```
